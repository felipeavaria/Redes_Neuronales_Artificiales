# 1. Rol de las Capas Ocultas de una ANN

(a) Se crean los conjuntos de datos de entrenamiento (X1, y1) y de prueba (X2, y2). Luego se visualiza el conjunto de datos de entrenamiento:

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
X1, y1 = make_moons(500, noise=0.20) #entrenamiento
X2, y2 = make_moons(500, noise=0.20) #pruebas
plt.gcf().clear() #limpia lo dibujado previamente en plt
plt.scatter(X1[:,0], X1[:,1], s=40, c=y1, cmap=plt.cm.Spectral)
#datos2=plt.scatter(X2[:,0], X2[:,1], s=40, c=y2, cmap=plt.cm.Spectral)

#graficar datos
plt.show()
```

Se crea un clasificador de una sola neurona: modelo_una_neurona. Usa como función de activación la Sigmoidal, función de pérdida la Binary Crossentropy (para clasificaciones binarias, lo cual es nuestro caso) y de optimizador el método rmsprop. El campo metrics indica que se evaluará el 'accuracy' al momento de poner el clasificador a prueba con evaluate(). Normalmente, evaluate() devuelve sólo el valor de la función de pérdida "loss" (la única útil para determinar el rendimiento del clasificador), pero al indicar metrics=['accuracy'] al compilar el modelo, le estamos diciendo a evaluate() que nos devuelva, además del valor de "loss", el accuracy (o exactitud) que logró el clasificador al momento de ponerse a prueba (porcentaje de clasificaciones correctas en el conjunto de pruebas).

```python
from keras.models import Sequential
from keras.layers import Dense, Activation

modelo_una_neurona = Sequential()
capa_una_neurona = Dense(units=1, input_dim=2)
modelo_una_neurona.add(capa_una_neurona)
modelo_una_neurona.add(Activation('sigmoid'))

modelo_una_neurona.compile(loss='binary_crossentropy',
              optimizer ='rmsprop',
              metrics=['accuracy'])
```

Se entrena el clasificador modelo_una_neurona con fit(). se elige batch_size = 50 (podria ser otro valor) y epoch=1 (debe ser 1) en cada ejecución de entranamiento. Esto es, del conjunto (X1, y1) se toman 50 datos, sin elegir al azar, con los cuales el clasificador se entrena, y al procesarlos, hace un forward pass y un backward pass (ajuste de pesos, creo). Como el conjunto de datos es de tamaño 500, se necesitan 500/50 = 10 iteraciones de procesamientos + forward pass/backward pass. Eso es lo que se hace en 1 epoch. Los valores de "loss" y "accuracy" del entrenamiento se almacenan.
Luego de entrenarse 1 vez con todo el conjunto de datos (1 epoch), se pone a prueba el clasificador con evaluate() y se extraen los valores de "loss" y "accuracy" resultantes de la evaluación del clasificador (recordar que entre menor es "loss" mejor es el clasificador; "accuracy" mide qué tan bien predijo del conjunto de prueba, pero no se usa para medir calidad).
El ciclo for de 10000 iteraciones lo que hace es repetir el entrenamiento y la prueba 10000 veces, almacenando los valores de loss y accuracy tanto del entrenamiento como de la prueba en cada iteración. Entrenar 10000 veces el mismo modelo usando epochs=1 es equivalente a entrenaar el modelo una sola usando epochs=10000, no se hizo de la última forma porque no supe cómo hacer que el modelo se sometiera a prueba (evaluate()) luego de terminar cada epoch. Necesitaba el "loss" y el "accuracy" de las pruebas para asi poder ir comparando el rendimiento de los modelos respecto a la cantidad de entrenamiento.

```python
lista_loss_entrenamiento_una_neurona = []
lista_acc_entrenamiento_una_neurona = []
lista_loss_prueba_una_neurona = []
lista_acc_prueba_una_neurona = []
for i in range(10000):
    #loss y acc del entrenamiento
    entrenamiento_una_neurona = modelo_una_neurona.fit(X1, y1, epochs=1, verbose=0, batch_size=50)
    lista_loss_entrenamiento_una_neurona.append(entrenamiento_una_neurona.history['loss'][0])
    lista_acc_entrenamiento_una_neurona.append(entrenamiento_una_neurona.history['acc'][0])
    #loss y acc de la prueba
    [loss_prueba,acc_prueba] = modelo_una_neurona.evaluate(X2, y2, verbose=0, batch_size=500)
    lista_loss_prueba_una_neurona.append(loss_prueba)
    lista_acc_prueba_una_neurona.append(acc_prueba)
```

Se intenta graficar (linea verde) la linea en representación del clasificador y su efecto sobre el conjunto de datos de prueba (X2).

```python
clases = modelo_una_neurona.predict(X2)

y_predicho = []
for clase_predicha in clases:
    if clase_predicha > 0.5:
        y_predicho.append(1)
    else:
        y_predicho.append(0)


plt.gcf().clear() #limpia lo dibujado previamente en plt
plt.scatter(X2[:,0], X2[:,1], s=40, c=y_predicho, cmap=plt.cm.Spectral)
dx = capa_una_neurona.get_weights()[0][0]
dy = capa_una_neurona.get_weights()[0][1]
n = -capa_una_neurona.get_weights()[1][0]
m = -(dy/dx)
print "Ecuacion de la recta del clasificador: "+str(m[0])+"x + "+str(n)
plt.plot([-1*m[0]+n, 1.5*m[0]+n], [-1,1.5], 'g')
#graficar datos
plt.show()
```

Ahora se grafica el conjunto de datos (X2) con su clasificación real (y2). Notar cómo el clasificador no logra separar las clases.

```python
plt.gcf().clear() #limpia lo dibujado previamente en plt
plt.scatter(X2[:,0], X2[:,1], s=40, c=y2, cmap=plt.cm.Spectral)
plt.plot([-1*m[0]+n, 1.5*m[0]+n], [-1,1.5], 'g')
#graficar datos
plt.show()
```

(b) Sin mayor razón elegí crear la segunda red (modelo_multicapas) con dos capas ocultas de 10 neuronas cada una, ambas con activación relu. Se inserta una última capa con 1 unidad y activación sigmoidal (neurona de salida). Se usan los mismos parámetros de compilación que en el modelo anterior.

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
from keras.optimizers import SGD

modelo_multicapas = Sequential()
modelo_multicapas.add(Dense(units=10, input_dim=2, activation='relu'))
#modelo_multicapas.add(Dropout(0.5)) #evita overfitting
modelo_multicapas.add(Dense(units=10, activation='relu'))
#modelo_multicapas.add(Dropout(0.5)) #evita overfitting, again
modelo_multicapas.add(Dense(1, activation='sigmoid'))

modelo_multicapas.compile(loss='binary_crossentropy',
                          optimizer='rmsprop',
                          metrics=['accuracy'])

```

El mismo procedimiento de entrenamiento que en el caso anterior. Se necesitan tanto los loss de entrenamiento como de prueba (los accuracy no de forma obligatoria, pero ayudan a visualizar el progreso de los modelos con el entrenamiento)

```python
lista_loss_entrenamiento_multicapas = []
lista_acc_entrenamiento_multicapas = []
lista_loss_prueba_multicapas = []
lista_acc_prueba_multicapas = []
for i in range(10000):
    #loss y acc del entrenamiento
    entrenamiento_multicapas = modelo_multicapas.fit(X1, y1, epochs=1, verbose=0, batch_size=50)
    lista_loss_entrenamiento_multicapas.append(entrenamiento_multicapas.history['loss'][0])
    lista_acc_entrenamiento_multicapas.append(entrenamiento_multicapas.history['acc'][0])
    #loss y acc de la prueba
    [loss_prueba,acc_prueba] = modelo_multicapas.evaluate(X2, y2, verbose=0, batch_size=500)
    lista_loss_prueba_multicapas.append(loss_prueba)
    lista_acc_prueba_multicapas.append(acc_prueba)
```

A continuación pongo en un gráfico los accuracy de pruebas y entrenamiento del modelo de una neurona y el de multicapas. Notar cómo los accuracy (entrenamiento y pruebas) del modelo de una neurona llegan a un tope y se mantienen en ese nivel aproximadamente todo el entrenamiento. Pasa lo mismo con los del modelo multicapas, pero éstos llegan a un tope mayor, bastante cercano a 1.0. Esto quiere decir que, el modelo multicapas logra clasificar correctamente mas del 95% de los datos de prueba, mientras que el clasificador de una sola neurona no logra pasar el 90% (ver segundo gráfico).

```python
plt.gcf().clear() #limpia lo dibujado previamente en plt
plt.plot(range(10001)[1:],lista_acc_entrenamiento_una_neurona, 'g', label='Entr. Una neurona')
plt.plot(range(10001)[1:],lista_acc_entrenamiento_multicapas, 'b', label='Entr. Multicapas')
plt.plot(range(10001)[1:],lista_acc_prueba_una_neurona, 'r', label='Prueba una neurona')
plt.plot(range(10001)[1:],lista_acc_prueba_multicapas, 'y', label='Prueba multicapas')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend(loc='lower right')
plt.axis([0,10000,0,1])
plt.show()
```

Mismo procedimmiento para los valores de "loss" de ambos modelos. Notar que aqui, el loss del modelo de una neurona logra descender hasta cierto valor límite, quedándose constante para los siguientes entrenamientos. O sea, a partir de cierto punto, ya no siguió mejorando al entrenar.
Con el modelo de multicapas, ocurrió que el loss de entrenamiento fue mejorando siempre con cada entrenamiento (o sea que, el modelo aprendió en cada iteración a clasificar de mejor forma el conjunto de entrenamiento), pero a partir de cierto punto, el loss de prueba empeoró y se mantuvo en esa tendencia. Esto se explica por el overfitting que el modelo multicapas experimentó a causa de la gran cantidad de entrenamiento con un mismo conjunto de datos.
Sin embargo, queda claro que con la cantidad de entrenamiento correcta se logra un mínimo en el loss de prueba aproximadamente con 150 epochs (el modelo aprende a generalizar bien y a clasificar nuevos datos). este loss mínimo, de aproximadamente 0.1, es mucho más inferior al mínimo en que se estanca el loss de prueba del modelo de una neurona (un poco menos de 0.3), demostrando experimentalmente que un modelo de varias capas resuelve con mayor satisfacción este problema de clasificación.
